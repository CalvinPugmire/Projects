```python
def get_matches_x ():
  df = pd.read_csv('CVL All Past Replays Pokemon Teams.csv')

  df_win = df['Winning Player\'s 6 Pokemon'].str.split(' / ', expand=True)
  df_los = df['Losing Player\'s 6 Pokemon'].str.split(' / ', expand=True)

  df_f = pd.DataFrame()

  for i in range(len(df_win)):
    c = np.random.choice([0, 1])
    if c == 0:
        df_concat = pd.concat([df_win.iloc[i], df_los.iloc[i], pd.Series([c])], axis=0)
    else:
        df_concat = pd.concat([df_los.iloc[i], df_win.iloc[i], pd.Series([c])], axis=0)
    df_f = pd.concat([df_f, df_concat], axis=1)

  df_f = df_f.transpose()
  df_f.columns = ['T1 P1', 'T1 P2', 'T1 P3', 'T1 P4', 'T1 P5', 'T1 P6',
                  'T2 P1', 'T2 P2', 'T2 P3', 'T2 P4', 'T2 P5', 'T2 P6',
                  'Winner']

  # print(df_f.to_string(max_rows=5000, max_cols=5000))
  #display(df_f)
  return df_f
get_matches_x()
```

```python
def get_matches_xy(matches=get_matches_x):
  dfm = matches()
  dfp = get_pokemon_x()

  to_clean = set()
  nan_count = 0
  outcomes = dfm['Winner']
  for index, row in dfm.iterrows():
    column_names = dfm.columns[:-1]

    for i in column_names:
      if row[i] is np.nan:
          nan_count += 1
      else:
          f_dfp = dfp[dfp['Name'] == row[i]]

      # Get the Stats of the Pokemon
      if f_dfp.index.size > 0:
          i1 = str(f_dfp['Type 1'].values[0])
          i2 = str(f_dfp['Type 2'].values[0])
          i3 = str(f_dfp['HP'].values[0])
          i4 = str(f_dfp['Defense'].values[0])
          i5 = str(f_dfp['Sp. Def'].values[0])
          i6 = str(f_dfp['Attack'].values[0])
          i7 = str(f_dfp['Sp. Atk'].values[0])
          i8 = str(f_dfp['Speed'].values[0])
          row[i] = i1+'/'+i2 +'/'+i3+'/'+i4+'/'+i5+'/'+i6+'/'+i7+'/'+i8
      else:
          row[i] = '21/21/0.0/0/0/0/0/0'
          """row[i] = "Not found:"+str(row[i])
          to_clean.add(str(row[i]))"""

  print(dfm.to_string(max_rows=10, max_cols=5000))
  print("Pokemon data to clean: ", to_clean)
  print(len(to_clean))
  print("Nan count: ", nan_count)

  df_win_1 = dfm['T1 P1'].str.split('/', expand=True)
  df_win_2 = dfm['T1 P2'].str.split('/', expand=True)
  df_win_3 = dfm['T1 P3'].str.split('/', expand=True)
  df_win_4 = dfm['T1 P4'].str.split('/', expand=True)
  df_win_5 = dfm['T1 P5'].str.split('/', expand=True)
  df_win_6 = dfm['T1 P6'].str.split('/', expand=True)
  df_los_1 = dfm['T2 P1'].str.split('/', expand=True)
  df_los_2 = dfm['T2 P2'].str.split('/', expand=True)
  df_los_3 = dfm['T2 P3'].str.split('/', expand=True)
  df_los_4 = dfm['T2 P4'].str.split('/', expand=True)
  df_los_5 = dfm['T2 P5'].str.split('/', expand=True)
  df_los_6 = dfm['T2 P6'].str.split('/', expand=True)
  df = pd.concat([df_win_1, df_win_2, df_win_3, df_win_4, df_win_5, df_win_6,
                  df_los_1, df_los_2, df_los_3, df_los_4, df_los_5, df_los_6,
                  outcomes], axis=1)

  print(df.to_string(max_rows=10, max_cols=5000))
  #display(df)
  return df
```

```python
def get_matches_scraped():
  df = pd.read_csv('scraped_data.csv')

  df.iloc[:, 0] = df.iloc[:, 0].str.replace('[\[\]\']', '', regex=True)
  df.iloc[:, 1] = df.iloc[:, 1].str.replace('[\[\]\']', '', regex=True)

  df_1 = df['team1'].str.split(', ', expand=True)
  df_2 = df['team2'].str.split(', ', expand=True)
  outcome = df['outcome']

  df_f = pd.DataFrame()

  df_f = pd.concat([df_1, df_2, outcome.astype('str')], axis=1)

  # df_f = df_f.transpose()
  # display(df_f)
  df_f.columns = ['T1 P1', 'T1 P2', 'T1 P3', 'T1 P4', 'T1 P5', 'T1 P6',
                  'T2 P1', 'T2 P2', 'T2 P3', 'T2 P4', 'T2 P5', 'T2 P6',
                  'Winner']

  # print(df_f.to_string(max_rows=5000, max_cols=5000))
  # display(df_f)
  return df_f
fff = get_matches_scraped()
```
